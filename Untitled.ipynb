```python
# Extracting train dataset

train = pd.read_csv('train.csv', nrows = 10_000_000)
```

```python
# Establishing a range for features based on observations above

train = train[train.iloc[:,1] < 600]
train = train[train.iloc[:,1] > 0]

train = train[train.iloc[:,3] < -71]
train = train[train.iloc[:,3] > -75]

train = train[train.iloc[:,4] > 39]
train = train[train.iloc[:,4] < 43]

train = train[train.iloc[:,5] < -71]
train = train[train.iloc[:,5] > -75]

train = train[train.iloc[:,6] > 39]
train = train[train.iloc[:,6] < 43]

train = train[train.iloc[:,7] <=4]
train = train[train.iloc[:,7] >0]
```

```python
# Shape

train.shape
```

```python
# Extracting test dataset

test = pd.read_csv('test.csv')
```

```python
test.head()
```

```python
# Target and features

y = train['fare_amount']
train_X = train.drop('fare_amount',axis=1)
```

```python
# Calculating difference between dropoffs and pickups on test dataset

def travel(test):
    test['Lat_Travel'] = (test['dropoff_latitude']-test['pickup_latitude'])
    test['Lon_Travel'] = (test['dropoff_longitude']-test['pickup_longitude'])
                                                              
    
travel(test)
```

```python
# Calculating difference between dropoffs and pickups on train dataset

def travel_2(train_X):
    train_X['Lat_Travel'] = (train_X['dropoff_latitude']-train_X['pickup_latitude'])
    train_X['Lon_Travel'] = (train_X['dropoff_longitude']-train_X['pickup_longitude'])
                                                              
    
travel_2(train_X)
```

```python
# Concat train and test datasets: all_data

test_X=test
train_X['train_set'] = 1
test_X['train_set'] = 0
all_data = pd.concat([train_X, test_X], axis=0, sort=False)
```

```python
all_data
```

```python
# Creating date features

all_data['year'] = all_data.pickup_datetime.apply(lambda k: k[:4])
all_data['month'] = all_data.pickup_datetime.apply(lambda k: k[5:7])
all_data['hour'] = all_data.pickup_datetime.apply(lambda k: k[11:13])
```

```python
all_data.head()
```

```python
# Date feature

all_data['pickup_datetime'] = all_data.pickup_datetime.apply(lambda k: dt.datetime.strptime(k[:10], '%Y-%m-%d'))
```

```python
# Data feature

all_data['weekday'] = all_data.pickup_datetime.apply(lambda k: k.weekday())
```

```python
all_data.head(3)
```

```python
# Drop features

all_data = all_data.drop(['key', 'pickup_datetime'], axis=1)
```

```python
all_data.info()
```

```python
# Changing features' types

all_data.year = all_data.year.astype('int')
all_data.hour = all_data.hour.astype('int')
```

```python
from sklearn.cluster import KMeans
```

```python
# Creating clusters for pickups

all_data_clus = all_data[['pickup_longitude', 'pickup_latitude']]
model = KMeans(n_clusters=6)
model.fit(all_data_clus)
pred = model.predict(all_data_clus)
    
all_data['pickup_cluster'] = pred
clusters = pd.get_dummies(all_data['pickup_cluster'],prefix='pickup_Cluster',drop_first=False)
all_data = pd.concat([all_data,clusters],axis=1).drop('pickup_cluster',axis=1)
```

```python
all_data.describe()
```

```python
# Creating clusters for dropoffs

all_data_clus = all_data[['dropoff_longitude', 'dropoff_latitude']]
model = KMeans(n_clusters=6)
model.fit(all_data_clus)
pred = model.predict(all_data_clus)
    
all_data['dropoff_cluster'] = pred
clusters = pd.get_dummies(all_data['dropoff_cluster'],prefix='dropoff_Cluster',drop_first=False)
all_data = pd.concat([all_data,clusters],axis=1).drop('dropoff_cluster',axis=1)
```

```python
# Separating train and test datasets

train = all_data[all_data.train_set == 1].drop(['train_set'], axis=1)
test = all_data[all_data.train_set == 0].drop(['train_set'], axis=1)
```

```python
# Using seaborne to make a pairplot

train['fare_amount'] = y
sns.pairplot(train[:500],vars=['Lat_Travel', 'Lon_Travel', 'fare_amount', 'passenger_count', 'weekday', 'year', 'hour'])
```

```python
# Splitting train and test data 

X = train.drop('fare_amount', axis=1)
y = train['fare_amount'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# Using Random Forest Regressor model to predict

rfg = RandomForestRegressor(n_estimators = 25, max_features = 20, max_depth = 25, min_samples_split=3,
                            min_samples_leaf=2, random_state=0)

rfg.fit(X_train, y_train)
predic = rfg.predict(X_test)

# Calculating error

e = np.sqrt(mean_squared_error(y_test,predic))
e
```

```python
# Plotting a barh graphic based on importance of each feature

feats = X.columns[:X.shape[1]]
imp = rfg.feature_importances_
ind = np.argsort(imp)

plt.title('Feature Importances')
plt.barh(range(len(ind)), imp[ind], align='center')
plt.yticks(range(len(ind)), [feats[i] for i in ind])
plt.xlabel('Relative Importance')
plt.show()
```

```python
# Using Random Forest Regressor model to predict on test dataset

rfg = RandomForestRegressor(n_estimators = 25, max_features = 20, max_depth = 25, min_samples_split=3,
                            min_samples_leaf=2, random_state=0)

rfg.fit(X,y)
pred = rfg.predict(test)
```
